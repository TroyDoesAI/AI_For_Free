```python
import requests
from base64 import b64encode
from PIL import Image
from IPython.display import display
```

```python
def toB64(imgUrl):
    return str(b64encode(requests.get(imgUrl).content))[2:-1]

url = "https://api.segmind.com/v1/ssd-img2img"
```
